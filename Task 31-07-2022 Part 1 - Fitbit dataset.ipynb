```python
#Task 8 - Explore over an internet that how much calories burn is required for a healthy person 
#         and find out how many healthy person we have in our dataset 

# 1800 calorie per day is expected to be burnt for any healthy person. Assumption, if calorie burnt per day > 1800, consider healthy.
# A new column healthIndex to be created with 1 if calorie spent > 1800, 0 if less than 1800
```

```python
def caloriecheck(a):
    if a>=1800:
        return 1
    else:
        return 0
```

```python
fitbit_df['HealthyDay'] = fitbit_df['Calories'].apply(caloriecheck)
```

```python
fitbit_df[['Id','HealthyDay']]
```

```python
health_df = fitbit_df.groupby(['Id'])[['HealthyDay','Calories']].sum() 
# sum of total calories in month
```

```python
health_df.head()
```

```python
health_df['LoggedDays']=fitbit_df.groupby('Id')['ActivityDate'].count().sort_values(ascending = False)
#create a new column to compute total number of logged days per ID
```

```python
health_df['AvgCalorie'] = health_df['Calories']/health_df['LoggedDays'] 
# Create a new column to find out average of calorie burnt
```

```python
health_df['AvgHealthIndex'] = health_df['HealthyDay']/health_df['LoggedDays'] 
# Create a new column to find out average of health index, here we divide healthy days by logged days.
```

```python
health_df.sort_values(by=['AvgCalorie','HealthyDay'],ascending=False).head()
```

```python
health_df.describe()
```

```python
#now we clasify health status if 
def healthstatus(a,b):
    if ((a > 1800.0) or (b >0.8)):
        return 'Healthy'
    else:
        return 'Unhealthy'
```

```python
health_df['healthStatus'] =health_df.apply(lambda health_df :healthstatus(health_df['AvgCalorie'],health_df['AvgHealthIndex']),axis =1)
```

```python
health_df.head()
```

```python
healthy_count = health_df[health_df['healthStatus'] == 'Healthy']
```

```python
#Task 9 How many person are not a regular person with respect to activity try to find out those 
```

```python
health_df.reset_index(inplace= True)
```

```python
def ActivityStatus(a):  # define a fuction to determine activity status if logged days is greater than = to 14 as 'Regular'
    if a >= 14 :
        return 'Regular'
    else:
        return 'Irregular'
```

```python
health_df['ActivityStatus']=health_df['LoggedDays'].apply(ActivityStatus)
#added a column 
```

```python
health_df[health_df['ActivityStatus'] == 'Regular'][['Id','LoggedDays','ActivityStatus']].sort_values(by='LoggedDays',ascending= False)
```
